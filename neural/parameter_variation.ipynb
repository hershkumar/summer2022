```python
import numpy as np 
from matplotlib import pyplot as plt
import gvar as gv

true_energy = 0.3098
energies = [0.30984178,0.30675635, .31607518, 0.30420572, .30097234, 0.31568897,0.3270345 ,0.32448924]
uncerts = [ 0.0051553384, 0.004757211, 0.008432422, 0.0060188673, 0.0104259625, 0.007303138,0.002082853153169509 ,0.0018503338410103848 ]
gvars = [gv.gvar(energies[i], uncerts[i]) for i in range(len(energies))]
errors = [(gvars[i] - true_energy)/true_energy for i in range(len(gvars))]

num_parameters = [ 92801, 65201, 40001, 22401, 9801, 2151, 416, 90]
log_params = np.log10(num_parameters)

frac_diff = [errors[i].mean for i in range(len(errors))]
uncerts = [errors[i].sdev for i in range(len(errors))]

plt.errorbar(log_params, frac_diff, yerr=uncerts, fmt='o')
plt.xlabel('log10(num parameters)')
plt.ylabel('Fractional difference')
plt.title("Fractional error vs. number of parameters")
# plot the true energy as a horizontal line
plt.axhline(y=0.0, color='r', linestyle='-')
plt.show()
```
