```python
import numpy as np
from matplotlib import pyplot as plt

# import the data from the two distributions
shallow = np.load("shallow_dist_ns10k_s1k.npy")
deep = np.load("deep_dist_ns20k_s1k.npy")
```

```python
# compute the mean and standard deviation of the two distributions
shallow_mean = np.mean(shallow)
shallow_std = np.std(shallow)
deep_mean = np.mean(deep)
deep_std = np.std(deep)

# plot the two on a histogram
plt.hist(shallow, bins=100, alpha=0.5, label='shallow')
plt.hist(deep, bins=100, alpha=0.5, label='deep')
# plot the mean and standard deviation of the two distributions
plt.axvline(shallow_mean, color='r', linestyle='dashed', linewidth=1)
plt.axvline(shallow_mean + shallow_std, color='r', linestyle='dashed', linewidth=1)
plt.axvline(shallow_mean - shallow_std, color='r', linestyle='dashed', linewidth=1)
plt.axvline(deep_mean, color='b', linestyle='dashed', linewidth=1)
plt.axvline(deep_mean + deep_std, color='b', linestyle='dashed', linewidth=1)
plt.axvline(deep_mean - deep_std, color='b', linestyle='dashed', linewidth=1)

plt.legend(loc='upper right')
plt.title("Histogram of shallow and deep Bosonic distributions")
plt.xlabel("Kinetic Energy")
plt.ylabel("Frequency")



# compute the widths of the two distributions
shallow_width = 2 * shallow_std
deep_width = 2 * deep_std
print("Shallow width: ", shallow_width)
print("Deep width: ", deep_width)


plt.show()
```
