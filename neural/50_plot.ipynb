```python
import numpy as np 
from matplotlib import pyplot as plt 

# read in the npz file
npzfile = np.load("-01_density_profiles.npz")

bins_nn = npzfile['x_bins_nn']
nn_x = npzfile['n_x_nn']
bins_exact = npzfile['x_bins_exact']
exact_x = npzfile['n_x_exact']


npzfile2 = np.load("05_density_profiles.npz")
bins_nn_plus = npzfile2['x_bins_nn']
nn_x_plus = npzfile2['n_x_nn']
bins_exact_plus = npzfile2['x_bins_exact']
exact_x_plus = npzfile2['n_x_exact']

```

```python
plt.plot(bins_nn, nn_x,'-o', markersize=2, color="darkred", label=r"Neural Network, $g = -0.01$")
plt.plot(bins_exact, exact_x,'-o', markersize=2, color="red", label=r"Analytic Solution, $g = -0.01$")

plt.plot(bins_nn_plus, nn_x_plus,'-o', markersize=2, color="blue", label=r"Neural Network, $g = 0.05$")
plt.plot(bins_exact_plus, exact_x_plus,'-o', markersize=2, color="royalblue", label=r"Analytic Solution, $g = 0.05$")

plt.title(r"$N = 50$ Local Density Profile, $g = -2\sigma$")
plt.xlabel("$x$")
plt.ylim(0, 1)
plt.grid()
plt.legend()
plt.ylabel("$n(x)/\int dx n(x)$")
plt.savefig("50_density_profiles.svg", format="svg")
plt.show()
```
